We are going to need other data that contain more data with more numbers but lets get the year they possessed nuclear weapon.


```python
countries_with_possession=df[df["nuclear_weapons_status"] == 3]
countries_year_possession = countries_with_possession.groupby("country_name")["year"].min().reset_index()
countries_year_possession
```

We can see that USA is the first country that possessed nuclear weapon in 1945 and North Korea is last country that have aquired nuclear weapon 
in 2006.
